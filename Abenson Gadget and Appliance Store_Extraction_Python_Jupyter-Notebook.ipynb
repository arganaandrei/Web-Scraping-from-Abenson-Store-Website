```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
```

```python
def get_title(soup):

    try:
        title = soup.find("span", attrs={"class":'base'})
        
        title_value = title.text

        title_string = title_value.strip()

    except AttributeError:
        title_string = ""

    return title_string


def get_price(soup):

    try:
        price = soup.find("span", attrs={"class":'price'}).string.strip()

    except AttributeError:
        price = ""
            
    return price

def get_rating(soup):

    try:
        rating = soup.find("i", attrs={"class":'action add-review'}).string.strip()
    
    except AttributeError:
        try:
            rating = soup.find("span", attrs={'class':'a-icon-alt'}).string.strip()
        except:
            rating = "no rating yet"	

    return rating

def get_discount(soup):
    try:
        discount = soup.find("span", attrs={'class':'price-box-discount'}).string.strip()

    except AttributeError:
        discount = ""

    return discount

def get_availability(soup):
    try:
        available = soup.find("div", attrs={'class':'stock available'})
        available = available.find("span").string.strip()

    except AttributeError:
        available = "Not Available"	

    return available
```

```python
if __name__ == '__main__':


    HEADERS = ({'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/107.0.0.0 Safari/537.36 Edg/107.0.1418.52', 'Accept-Language': 'en-US, en;q=0.5'})


    URL = "https://www.abenson.com/computers-gadget/laptops-computers/computer-hardware-type/laptop.html"

    webpage = requests.get(URL, headers=HEADERS)

    soup = BeautifulSoup(webpage.content, "html.parser")

    links = soup.find_all("a", attrs={'class':'product-item-link'})

    links_list = []

    for link in links:
            links_list.append(link.get('href'))

    d = {"title":[], "price":[], "rating":[], "discount":[],"availability":[]}
    
    for link in links_list:
        new_webpage = requests.get(link, headers=HEADERS)

        new_soup = BeautifulSoup(new_webpage.content, "html.parser")

        d['title'].append(get_title(new_soup))
        d['price'].append(get_price(new_soup))
        d['rating'].append(get_rating(new_soup))
        d['discount'].append(get_discount(new_soup))
        d['availability'].append(get_availability(new_soup))

    
    abenson_df = pd.DataFrame.from_dict(d)
    abenson_df['title'].replace('', np.nan, inplace=True)
    abenson_df = abenson_df.dropna(subset=['title'])
    abenson_df.to_csv("abenson_df.csv", header=True, index=False)
```
